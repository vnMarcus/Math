### Pyplot API vs Object-Oriented API
* Pyplot API -> Quickly
* OO API -> Advanced

Pylot API

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "blue", linestyle="dashed", label="Sin(x)");
plt.plot(x, np.cos(x), color = "red", linestyle="dashed", label="Cos(x)");
plt.xlabel("Bien x")
plt.ylabel("Fx")
plt.title("Sin(x) and Cos(x)")
# plt.xlim(0,4)
# plt.ylim(-0.6, 0.6)
# plt.axis([xmin, xmax, ymin, ymax])
plt.axis("equal")
plt.legend();
```

# OO API

```python
# Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# Setup plot 
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, y)
ax.set(title="A simple plot",
      xlabel= "x-axis", 
      ylabel="y-axis");
```

### Scatter 

```python
# Pylot API
plt.scatter(x, np.exp(x)) # y = e^(x)
```

```python
# OO API
# Prepare New Data
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
size = 1000 * rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s = size, c = color, cmap ='viridis', alpha = 0.3);
fig.colorbar(img1);
```

# Bar 
* Vertical
* Horizontal

```python
# Prepare data from python dictionary
soft_drink_prices = {
    "Coke":10,
    "Pessi":12,
    "Sprite":8
}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title="Bach Hoa Xanh's soft drink prices", ylabel="Price ($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()));
ax.set(title="Bach Hoa Xanh's soft drink prices", xlabel="Price ($)");
```

# Histogram

```python
# Prepare data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
student_height[:10]
```

```python
# pyplot API
plt.hist(student_height, bins= 30);
```

```python
# Subplots
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10,5))
# Plot data to each axis 
ax1.plot(x, x/2)
ax1.plot(x, x*2)

ax2.scatter(np.random.random(10),np.random.random(10) )
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values())

ax4.hist(student_height, bins= 30)
```

```python
# Option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize =(10, 5))
ax[0,0].plot(x, x/2)
ax[0,0].plot(x, x*2)

ax[0,1].scatter(np.random.random(10),np.random.random(10) )
ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values())

ax[1,1].hist(student_height, bins= 30)
```
